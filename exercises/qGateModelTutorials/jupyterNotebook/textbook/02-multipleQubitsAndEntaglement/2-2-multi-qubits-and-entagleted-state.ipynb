```python
from qiskit import QuantumCircuit, Aer, assemble
from math import pi
import numpy as np
from qiskit.visualization import plot_histogram, plot_bloch_multivector
```

```python
qc = QuantumCircuit(3)
# Apply H-gate to each qubit:
for qubit in range(3):
    qc.h(qubit)
# See the circuit:
qc.draw(output='mpl')
```

```python
# Let's see the result
svsim = Aer.get_backend('statevector_simulator')
qobj = assemble(qc)
final_state = svsim.run(qobj).result().get_statevector()

# In Jupyter Notebooks we can display this nicely using Latex.
# If not using Jupyter Notebooks you may need to remove the 
# array_to_latex function and use print(final_state) instead.
from qiskit_textbook.tools import array_to_latex
array_to_latex(final_state, pretext="\\text{Statevector} = ")
```

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.x(1)
qc.draw(output='mpl')
```

```python
usim = Aer.get_backend('unitary_simulator')
qobj = assemble(qc)
unitary = usim.run(qobj).result().get_unitary()
```

```python
from qiskit_textbook.tools import array_to_latex
array_to_latex(unitary, pretext="\\text{Circuit = }\n")
```

```python
# applying a gate to only one qubit at a time
qc = QuantumCircuit(2)
qc.x(1)
qc.draw(output='mpl')
```

```python
# Simulate the unitary
usim = Aer.get_backend('unitary_simulator')
qobj = assemble(qc)
unitary = usim.run(qobj).result().get_unitary()
# Display the results:
array_to_latex(unitary, pretext="\\text{Circuit = } ")
```

```python
# 3.1 The CNOT-Gate
qc = QuantumCircuit(2)
# Apply CNOT
qc.cx(0,1)
# See the circuit:
qc.draw(output='mpl')
```

```python
qc = QuantumCircuit(2)
# Apply the H-gate to the first:
qc.h(0)
qc.draw(output='mpl')
```

```python
# Let's the result:
svsim = Aer.get_backend('statevector_simulator')
qobj = assemble(qc)
final_state = svsim.run(qobj).result().get_statevector()
# Print the statevector neatly:
array_to_latex(final_state, pretext="\\text{Statevector = }")
```

```python
#Let's see what happends when we apply the CNOT gate:
qc = QuantumCircuit(2)
# Apply H-gate to the first:
qc.h(0)
# Apply a CNOT:
qc.cx(0,1) # qubit 0th is the control, and qubit 1th is the target 
qc.draw(output='mpl')
```

```python
# Let's get the result:
qobj = assemble(qc)
result = svsim.run(qobj).result()
# Print the statevector neatly:
final_state = result.get_statevector()
array_to_latex(final_state,pretext="\\text{Statevector = }")
```

```python
plot_histogram(result.get_counts())
```

```python
plot_bloch_multivector(final_state)
```

```python
from qiskit.visualization import plot_state_qsphere
plot_state_qsphere(final_state)
```
