```python
from qiskit import QuantumCircuit
from qiskit.circuit import Gate
from math import pi

qc = QuantumCircuit(2)
c = 0
t = 1
```

```python
# 1 Making a Controlled-Z from a CNOT
qc.cz(c,t)
qc.draw(output='mpl')
```

```python
qc1 = QuantumCircuit(2)
# also a controlled-Z
qc1.h(t)
qc1.cx(c,t)
qc1.h(t)
qc1.draw(output='mpl')
```

```python
qc2 = QuantumCircuit(2)
#  a controlled-Y
qc2.sdg(t)
qc2.cx(c,t)
qc2.s(t)
qc2.draw(output='mpl')
```

```python
qc3 = QuantumCircuit(2)
# a controlled-H
qc3.ry(pi/4,t)
qc3.cx(c,t)
qc3.ry(-pi/4,t)
qc3.draw(output='mpl')
```

```python
# 2 Swapping Qubits
a = 0
b = 1
```

```python
qc4 = QuantumCircuit(2)
# swaps states of qubits a and b
qc4.swap(a,b)
qc4.draw(output='mpl')
```

```python
qc5 = QuantumCircuit(2)
# swap a 1 from a to b
qc5.cx(a,b) # copies 1 from a to b
qc5.cx(b,a) # uses the 1 on b to rotate the state of a to 0 
qc5.draw(output='mpl')
```

```python
# sawp a q from b to a
qc5.cx(b,a) # copies 1 from b to a
qc5.cx(a,b) # copies the 1 on a to rotate the state of b to 0
qc5.draw(output='mpl')
```

```python
qc6 = QuantumCircuit(2)
qc6.cx(b,a)
qc6.cx(a,b)
qc6.cx(b,a)
qc6.draw(output='mpl')
```

```python
qc7 = QuantumCircuit(2)
# swaps states of qubits a and b
qc7.cx(a,b)
qc7.cx(b,a)
qc7.cx(a,b)
qc7.draw(output='mpl')
```

```python
# 3 Controlled Rotations
qc8 = QuantumCircuit(2)
theta = pi # theta can be anything (pi chosen arbitrarily)
qc8.ry(theta/2,t)
qc8.cx(c,t)
qc8.ry(-theta/2,t)
qc8.cx(c,t)
qc8.draw(output='mpl')
```

```python
A = Gate('A', 1, [])
B = Gate('B', 1, [])
C = Gate('C', 1, [])
alpha = 1 # arbitrarily define alpha to allow drawing of circuit
```

```python
qc9 = QuantumCircuit(2)
qc9.append(C, [t])
qc9.cz(c,t)
qc9.append(B, [t])
qc9.cz(c,t)
qc9.append(A, [t])
qc9.p(alpha,c)
qc9.draw(output='mpl')
```

```python
# 4 The Toffoli Gate
qc10 = QuantumCircuit(3)
a = 0
b = 1
t = 2
# Toffoli with control qubits a and b and target t
qc10.ccx(a,b,t)
qc10.draw(output='mpl')
```

```python
qc11 = QuantumCircuit(3)
qc11.cp(theta,b,t)
qc11.cx(a,b)
qc11.cp(-theta,b,t)
qc11.cx(a,b)
qc11.cp(theta,a,t)
qc11.draw(output='mpl')
```

```python
qc12 = QuantumCircuit(3)
qc12.ch(a,t)
qc12.cz(b,t)
qc12.ch(a,t)
qc12.draw(output='mpl')
```

```python
# 5 Arbitrary rotations from H and T
qc13 = QuantumCircuit(1)
qc13.t(0) # T gate on qubit 0
qc13.draw(output='mpl')
```

```python
qc14 = QuantumCircuit(1)
qc14.h(0)
qc14.t(0)
qc14.h(0)
qc14.draw(output='mpl')
```

```python
qc15 = QuantumCircuit(1)
qc15.h(0)
qc15.t(0)
qc15.h(0)
qc15.t(0)
qc15.draw(output='mpl')
```

```python
qc16 = QuantumCircuit(1)
qc16.t(0)
qc16.h(0)
qc16.t(0)
qc16.h(0)
qc16.draw(output='mpl')
```
